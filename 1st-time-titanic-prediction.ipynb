**DEFINIZIOE DATASET**

```python
train = pd.read_csv("/kaggle/input/titanic/train.csv")
test = pd.read_csv("/kaggle/input/titanic/test.csv")
```

**PULIZIA DEI DATI NAN DEL DATASET**

```python
valNulliTrain = train.isnull().sum()
print(valNulliTrain )
mediana = train["Age"].median()
print(f"il valore della mediana è {mediana}")
train["Age"] = train["Age"].fillna(mediana)
valNulliAgg = train.isnull().sum()
print(valNulliAgg)
```

```python
valNulliTest = test.isnull().sum()
print(valNulliTest)
mediana = test["Age"].median()
print(f"il valore della mediana è {mediana}")
test["Age"] = test["Age"].fillna(mediana)
valNulliAgg = test.isnull().sum()
print(valNulliAgg)
```

**CONVERSIONE DEI DATI CATEGORICI IN NUMERICI**

```python
train["Sex"] = train["Sex"].replace({"male": 1, "female": 0})
test["Sex"] = test["Sex"].replace({"male": 1, "female": 0})
print(train)
```

```python
X_train = train[["Pclass","Sex", "Age"]]
y_train= train["Survived"]

X_test = test[["Pclass","Sex", "Age"]]
```

**ALLENAMENTO DEL MODELLO**

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

**PREDIZIONI MODELLO**

```python
y_pred = model.predict(X_test)
```

**CONVERSIONE PROBABILITA' PREDETTE IN VALORI BINARI**

```python
y_pred_binary = (y_pred > 0.5).astype(int)
```

**CREAZIONE DATASET SUBMISSION**

```python
predictions = pd.DataFrame({
    "PassengerId": test["PassengerId"],  
    "Survived":y_pred_binary
})

predictions.to_csv("predictions.csv", index=False)  
print(predictions)
```
